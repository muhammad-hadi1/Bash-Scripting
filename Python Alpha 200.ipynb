```python
#List examples
list_a=['apple', 'banana', 'cherry']

list_b=[1, 2, 3, 4, 5]

print(list_a)

print(list_b)
```

```python
#nested List examples
list_a=['apple', 'banana', 'cherry']

list_b=[1, 2, 3, 4, 5]

list_c=[True, False, True]

all_list=list_a + list_b + list_c

print(all_list)
```

```python
# List extend method
com_list=[]
one_list=['a', 'b', 'c']
com_list.extend(one_list)
com_list
```

```python
#A slightly easier method to add two lists
my_list=[1, 2, 3]
new_list=[4, 5, 6]
combo_list=my_list + new_list
combo_list
```

```python
#Sorting out a list
alpha_list=['d', 'a', 'c', 'b']
alpha_list.sort()
alpha_list
```

```python
#Sorting out a list
list_num=[3 , 2 , 5 , 4 , 1 , 9 , 13 , 11]
list_num.sort()
list_num
```

```python
#A none list
list_num=[3 , 2 , 5 , 4 , 1 , 9 , 13 , 11]
sorted_list=list_num.sort()
sorted_list
print(sorted_list)
```

```python
#Slicing a list
alpha_list=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
sliced_list=alpha_list[2:7]
sliced_list
```
