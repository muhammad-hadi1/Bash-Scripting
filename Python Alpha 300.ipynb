```python
#if statement with else
Ali_age = 18
Ahmed_age = 20
if Ali_age > Ahmed_age:
    print("Ali is older than Ahmed")
else:
    print("Ahmed is older than Ali")    
```

```python
#if statement with elif
value1 = 9
if value1 > 10:
    print("value1 is greater than 10")
elif 10<= value1 <=20:
    print("value1 is between 10 and 20")
else:
    print("value1 is less than 10")
```

```python
#None examples for list, tuple, string and nothing
empty_list = []
empty_tuple = ()
empty_string = ""
nothing = None
if empty_list == []:
    print("It's an empty list!")
if empty_tuple:
    print("It's not an empty tuple!")
if not empty_string:
    print("This is an empty string!")
if not nothing:
    print("Then it's nothing!")
```
