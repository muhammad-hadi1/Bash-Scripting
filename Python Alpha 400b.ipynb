```python
#while loop example
a = 0
while a < 2:
    print(a)
    a += 1
```

```python
#while loop example with break
i = 0
while i < 5:
    print(i)
    if i == 5:
        break
    i += 1
```

```python
#while loop example with flow control
i=0
while i < 10:
    if i == 3:
        i +=1
        continue
    print(i)
    if i ==5:
        break
    i += 1
```

```python
#else is for in loops complex example
list = [1,2,3]
for item in list:
    if item == 8:
        break
else:
    print("item not found")    
```
